```
def print_sultan(result, return_code =False):
  print(*result.stdout, sep='\n')
  #print(result.stdout) # the stdout

  if len(result.stderr)>0: print(result.stderr)
  if len(result.traceback)>0: print(result.traceback)
  if return_code: print(result.rc)
```

```
#!pip install --upgrade sultan
from sultan.api import Sultan

with Sultan.load() as s:    
    result = s.ls("-l").run()
    print_sultan(result)
```

# Moving files in drive
Manual Organization of drive files

```
import os
def mysearch_files(win_path, filetype):
  base_path = drivepath_win2colab(win_path)
  #print(base_path)
  os.chdir(base_path)

  with Sultan.load(logging=False) as s:
    resultout = s.ls().pipe().grep(f'\.{filetype}$').run().stdout
    for i, c in enumerate(resultout):
      print(i, c)
  return resultout

def drivepath_win2colab(win_path):
  colabpath = win_path.replace("\\","/")
  colabpath = colabpath.replace("C:/Users/jason/Google Drive", "/content/drive/My Drive")
  colabpath+="/"
  return colabpath

## Testing  drivepath_win2colab
# win_path = r'C:\Users\jason\Google Drive'
# print(drivepath_win2colab(win_path))
```

```
# define path and filetype to search
filetype = "pdf"

win_path = r"""
C:\Users\jason\Google Drive
""".replace("\n","").replace("\t","")

resultout=mysearch_files(win_path, filetype)
```

```
# define file to move and target path
myfile = resultout[0]
target_path = r"""
C:\Users\jason\Google Drive\aDrive08\Corporeal\documents and certifications\civiles, USA
""".strip("\n")

target_path = drivepath_win2colab(target_path)
path1 = f'"{base_path}{myfile}"'
path2 = f'"{target_path}{myfile}"'

with Sultan.load(logging=False) as s:
  s.mv(path1, path2).run()

# update results  
resultout=mysearch_files(win_path, filetype)
```
